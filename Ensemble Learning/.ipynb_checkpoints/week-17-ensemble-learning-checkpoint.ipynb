## **ðŸ”¸ Boosting**

### LSBoost from scratch

<img src="imgs/lsboost.png" alt="string indexing" width=300 align="center" />

```python
class LSBoost:
    def __init__(self, n_estimators=100, learning_rate=0.1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.models = []

    def fit(self, X, y):
        H = np.zeros(len(y)) + np.mean(y)
        for _ in range(self.n_estimators):
            r = y - H
            model = DecisionTreeRegressor(max_depth=1)
            model.fit(X, r)
            h = model.predict(X)
            H += self.learning_rate * h
            self.models.append(model)
            
    def predict(self, X):
        H = np.zeros(len(X))
        for j in range(self.n_estimators):
            h = self.models[j].predict(X)
            H += self.learning_rate * h
        return H


# Create LSBoost instance
lsboost = LSBoost(n_estimators=150, learning_rate=0.5)

# Fit the model
lsboost.fit(X_train3, y_train3[:, 0])

# Make predictions
y_hat = lsboost.predict(X_test3)
print(r2_score(y_hat, y_test3[:, 0]))

plt.scatter(X_train3, y_train3, c='red', marker='x')
plt.scatter(X_test3, y_test3, c='blue', marker='x')

xp = np.linspace(X_train3.min(), X_train3.max(), 100)[:, None]
yp = lsboost.predict(xp)
plt.plot(xp, yp, c='gold', linewidth=3)

plt.legend(['Train', 'Test']);
```

## **ðŸ”¸ Stacking**

### Stacking for Classification

```python
model = StackingClassifier(estimators=[('knn', KNeighborsClassifier()),
                                       ('logreg', LogisticRegression(random_state=42)),
                                       ('tree', DecisionTreeClassifier(max_depth=4))],
                           final_estimator=RandomForestClassifier(n_estimators=16, random_state=42)
                          )
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Stacking for Regression

```python
model = StackingRegressor(estimators=[('knn', KNeighborsRegressor(n_neighbors=50)),
                                       ('rf', RandomForestRegressor(n_estimators=5, random_state=42)),
                                       ('tree', DecisionTreeRegressor(max_depth=3))],
                           final_estimator=LinearRegression()
                          )
model.fit(X_train3, y_train3.ravel())
print(model.score(X_test3, y_test3))

plt.scatter(X_train3, y_train3, c='red', marker='x')
plt.scatter(X_test3, y_test3, c='blue', marker='x')

xp = np.linspace(X_train3.min(), X_train3.max(), 100)[:, None]
yp = model.predict(xp)
plt.plot(xp, yp, c='gold', linewidth=3)

plt.legend(['Train', 'Test']);
```
